# Descriptive Statistics and Visualizations

**1. For the "Median value of owner-occupied homes" provide a boxplot**

```python
# Create the boxplot
ax = sns.boxplot(y='MEDV', data=boston_df)

# Calculate the median
median_value = boston_df['MEDV'].median()

# Set the title with median value
ax.set_title(f'Owner-Occupied Homes (Median: {median_value})')

# Add the median value to the boxplot
ax.text(0.95, median_value, f'Median: {median_value}', ha='right', va='center', color='red')

# Show the plot
plt.show()
```

**2. Provide a  bar plot for the Charles river variable**

```python
ax2 = sns.countplot(x='CHAS', data=boston_df)
ax2.set_title('Number of homes near the Charles River')

# Customize color scheme
colors = ['#FF7F0E', '#1F77B4']  # Set your desired colors for the two categories

# Assign colors to the bars
for i, patch in enumerate(ax2.patches):
    patch.set_facecolor(colors[i])

# Display the plot
plt.show()
#In this code, the colors variable contains a list of two colors that you can customize according to your preferences. The for loop assigns the colors to each bar in the countplot based on the order they appear in the plot. The ax2.legend() function creates a legend with the specified labels, and ax2.legend(title='Location') sets a title for the legend. Finally, plt.show() displays the plot with the legend.
```

**3. Provide a boxplot for the MEDV variable vs the AGE variable. (Discretize the age variable into three groups of 35 years and younger, between 35 and 70 years and 70 years and older)**

```python
boston_df.loc[(boston_df['AGE'] <= 35), 'Age_Group'] = '35 years and younger'
boston_df.loc[(boston_df['AGE'] > 35) & (boston_df['AGE'] < 70), 'Age_Group'] = 'between 35 and 70 years'
boston_df.loc[(boston_df['AGE'] >= 70), 'Age_Group'] = '70 years and older'
ax3 = sns.boxplot(x = 'MEDV', y = 'Age_Group', data = boston_df)
ax3.set_title('Median value of owner-occupied homes per Age Group')
```

**4. Provide a scatter plot to show the relationship between Nitric oxide concentrations and the proportion of non-retail business acres per town. What can you say about the relationship?**

```python
ax4 = sns.scatterplot(y='NOX', x='INDUS', data=boston_df)

# Calculate the correlation coefficient
corr = np.corrcoef(boston_df['INDUS'], boston_df['NOX'])[0, 1]

# Set the title
ax4.set_title('Nitric oxide concentration per proportion \n of non-retail business acres per town')

# Add legend with correlation
legend_labels = [f'Correlation: {corr:.2f}']
ax4.legend(title='Statistics', labels=legend_labels)

# Display the plot
plt.show()
```

# Tests for Statistics

**1. Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)**

## Hypothesis:

**Null Hypothesis:** There's no significant difference in median value between houses bounded and not bounded by the Charles River

**Alternative Hypothesis:** There's a significant difference in median value between houses bounded and not bounded by the Charles River

```python
boston_df.loc[(boston_df['CHAS'] == 0), 'CHAS_T'] = 'FAR'
boston_df.loc[(boston_df['CHAS'] == 1), 'CHAS_T'] = 'NEAR'
boston_df.head()
```

```python
scipy.stats.ttest_ind(boston_df[boston_df['CHAS_T'] == 'FAR']['MEDV'], 
                      boston_df[boston_df['CHAS_T'] == 'NEAR']['MEDV'], equal_var = True)
```

Based on a p-value of less than 0.05, we reject the Null Hypothesis, indicating no statistical difference in median value between houses near and far from the Charles River.

**2. Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)**

**Hypothesis**

**Null Hypotesis:** There isn't statistical difference in Median values of houses (MEDV) for each proportion of owner occpied units built prior to 1940

**Alternative Hypothesis:** There is statistical difference in Median values of houses (MEDV) for each proportion of owner occpied units built prior to 1940

```python
from statsmodels.formula.api import ols
lm = ols('MEDV ~ AGE', data = boston_df).fit()
table = sm.stats.anova_lm(lm)
print(table)
```

Given a p-value of less than 0.05, we reject the Null Hypothesis. Therefore, there is a statistical difference in median values of houses (MEDV) for each proportion of owner-occupied units built prior to 1940.

**3. Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation)**

**Null Hypothesis:** Nitric Oxide concentration is not correlated with the proportion of non-retail business acres per town

**Alternative Hypothesis:** Nitric Oxide concentration is correlated with the proportion of non-retail business acres per tow

```python
scipy.stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])
```

Given a Pearson coefficient of **0.76365** and a p-value less than **0.05**, we reject the Null Hypothesis. This confirms a positive correlation between nitric oxide concentration and the proportion of non-retail business acres per town.

The positive relationship is further supported by the Scatter Plot (Question **4**).

**4. What is the impact of an additional weighted distance to the five Boston employment centres on the median value of owner occupied homes? (Regression analysis)**

```python
x = boston_df['DIS']
y = boston_df['MEDV']

x = sm.add_constant(x)

model = sm.OLS(y, x).fit()
prediction = model.predict(x)

model.summary()
```
